```python
import numpy as np
import pandas as pd

songs = pd.read_csv('SongCSV.csv', decimal =',')

songs.head(5)
```

```python
# songs.drop(columns = "KeySignatureConfidence", inplace = True)
# songs.drop(columns = "TimeSignatureConfidence", inplace = True)

songs["Song_Hotness"] = songs["Song_Hotness"].astype(float)
songs["Danceability"] = songs["Danceability"].astype(float)
songs["Loudness"] = songs["Loudness"].astype(float)
songs["ArtistLatitude"] = songs["ArtistLatitude"].astype(float)
songs["ArtistLongitude"] = songs["ArtistLongitude"].astype(float)
songs["Tempo"] = songs["Tempo"].astype(float)
songs["Duration"] = songs["Duration"].astype(float)
songs["Artist_Familiarity"] = songs["Artist_Familiarity"].astype(float)
songs["Artist_Hotness"] = songs["Artist_Hotness"].astype(float)
songs["Artist_Terms"] = songs["Artist_Terms"].astype(float)
```

```python
for index, row in songs.iterrows():
    songs.ArtistName[index] = songs.ArtistName[index].lstrip("b'").rstrip("'")
    songs.SongID[index] = songs.SongID[index].lstrip("b'").rstrip("'")
    songs.AlbumName[index] = songs.AlbumName[index].lstrip("b'").rstrip("'").rstrip('"')
    songs.ArtistID[index] = songs.ArtistID[index].lstrip("b'").rstrip("'")
    songs.ArtistLocation[index] = songs.ArtistLocation[index].lstrip("b'").rstrip("'")
    songs.Title[index] = songs.Title[index].lstrip("b'").rstrip("'")
```

```python
songs.set_index("SongNumber", inplace = True)
```

```python
songs.shape
```

```python
song_year = songs.Year[songs.Year!= 0]
song_year
```

```python
song_less = songs[['ArtistName', 'Duration', "Tempo", 'Year',
       'AlbumName','ArtistLocation', 'Artist_Familiarity', 'Artist_Hotness' ]]
song_less
```

```python
from math import floor

def year_floor(x):
    return floor(x / 10)*10

def duration_floor(x):
    return floor(x / 10)*10 / 60

def artist_hotness(x):
    return floor(x * 50)/10

def tempo(x):
    return floor(x / 10) * 10

```

```python
song_less_year0 = song_less[song_less['Year']!=0]
song_less_year0 = song_less[song_less['Year']>1950]
song_less_year0 = song_less_year0[song_less['Tempo']!=0]
song_less_year0 = song_less_year0[song_less['Artist_Hotness']!=0]
song_less_year0 = song_less_year0[song_less['Artist_Familiarity']!=0]
song_less_year0 = song_less_year0[song_less['Duration']<800]

#song_less_y0 = song_less_year0.copy()

song_less_y0.Year = song_less_year0.Year.apply(lambda x: year_floor(x))
song_less_y0.Duration = song_less_year0.Duration.apply(lambda x: duration_floor(x))
song_less_y0.Artist_Hotness = song_less_year0.Artist_Hotness.apply(lambda x: artist_hotness(x))
song_less_y0.Tempo = song_less_year0.Tempo.apply(lambda x: tempo(x))

# song_less_y0 = song_less_year0.copy()
# song_less_y0
```

```python
# Import the library
import plotly.express as px

# Load the iris dataset provided by the library
df = song_less_y0

# Create the chart:
#fig = px.parallel_coordinates(df)

#fig = px.parallel_categories(df)

fig = px.parallel_coordinates(df, color= 'Artist_Familiarity',
                              dimensions=['Artist_Hotness', 'Tempo','Duration', 'Year'],
                              color_continuous_scale=px.colors.cyclical.HSV,
                              color_continuous_midpoint=0.75,
                              template='plotly_dark',
                             title = 'The MillSE PROJECT')
#     #color= "Artist Location", 
    #labels={"species_id": "Species","sepal_width": "Sepal Width", "sepal_length": "Sepal Length", "petal_width": "Petal Width", "petal_length": "Petal Length", },
    #range_color = [7])
    #color_continuous_scale=px.colors.diverging.Tealrose,
    #color_continuous_midpoint=2)

# Hide the color scale that is useless in this case
#fig.update_layout(coloraxis_showscale=False)


# Show the plot
fig.show()
```
